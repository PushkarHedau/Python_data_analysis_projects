## Separating Categorical and Numerical Variables

```python
# Seprating out the categorical and continuous variables
def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:                # looping on the number of columns
        if df[column].dtype== object:     
                                                   
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(df)         # Calling the function


table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].set_title("Count of origin")                  # Set the title of the graph
percentage = df["origin"].value_counts()         # value_counts: returns the count of values in series: return series
labels = list(df["origin"].value_counts().index) # Converting the index of value counts into list

sns.countplot(x = df["origin"], ax = ax[0])
plt.pie(percentage,labels = labels, autopct= "%0.2f%%")  # autopct: to print the percentage on pie wedge. 2 decimal points used 

plt.show()
```

```python

fig, ax = plt.subplots(1, 2, figsize= (15, 7))               # Creating the space for the 2 images, first argument in subplots is rows and second is columns and the figure size
sns.histplot(x = df["mpg"], ax =ax[1]) # ax[1] means image will plot at the second space
sns.boxplot(x = df["mpg"], ax = ax[0]) # ax[0] means image will plot at the second space
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def bar_count(ax, data):
    """
    Given a matplotlib axis object and a dataset, add count values as text labels to each bar in a bar chart.
    """
    # Get the total count of values in the dataset
    count = len(data)
    
    # Add the count values as text labels
    for bar in ax.patches:
        height = bar.get_height()
        ax.annotate(f"{height}",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center',
                    va='bottom')

    # Show the plot
    plt.show()

```

```python
fig, ax = plt.subplots(figsize = (10, 7))
sns.countplot(x = df["cylinders"]) 
bar_count(ax,df)
plt.show()
```
